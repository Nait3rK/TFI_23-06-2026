```python
import networkx as nx
grafod = nx.DiGraph()
grafod.add_nodes_from([1,2,3,4,5,6,7,8])
grafod.add_edges_from([(1,2),(1,3),(2,3),(2,4),(3,4),(3,5),(4,5),(4,6),(5,6),(5,7),(6,7),(6,8),(7,8),(7,1),(8,1),(8,2)])

colores = ['red','skyblue', 'gold', 'pink', 'orange', 'blue', 'purple', 'green']

nx.draw(grafod, with_labels=True, font_weight='bold', pos=nx.spectral_layout(grafod), node_color=colores, node_size=800)
```
